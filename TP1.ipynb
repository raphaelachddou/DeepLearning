# A regression Problem

```python
[X_train, Y_train] = generate_dataset_regression(1000, 20)
```

```python
import numpy as np
import numpy.linalg as la
 

def py_ang(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'    """
    cosang = np.dot(v1, v2)
    sinang = la.norm(np.cross(v1, v2))
    return np.arctan2(sinang, cosang)
def preprocessing(y):
    
    u0,v0,u1,v1,u2,v2 = y
    (xc,yc) = ((1./3)*(u0+u1+u2),(1./3)*(v0+v1+v2))
    
    ax = np.array([0.1,0])
    v = np.array([u0-xc,v0-yc])
    pi0 = py_ang(ax, v)
    if (v[1]<0):
        pi0 = 2*np.pi - pi0
    v = np.array([u1-xc,v1-yc])
    pi1 = py_ang(ax, v)
    if (v[1]<0):
        pi1 = 2*np.pi - pi1
    v = np.array([u2-xc,v2-yc])
    pi2 = py_ang(ax, v)
    if (v[1]<0):
        pi2 = 2*np.pi - pi2
    pi = [pi0,pi1,pi2]
    ind = sorted(range(len(pi)), key=lambda k: pi[k])
    res = []
    for i in ind :
        if i==0:
            res.append(u0)
            res.append(v0)
        if i==1:
            res.append(u1)
            res.append(v1)
        if i==2:
            res.append(u2)
            res.append(v2)
    return(np.array(res))
#print(Y_train[2])
#preprocessing(Y_train[2])
for i in range(Y_train.shape[0]):
    y = Y_train[i].copy()
    Y_train[i] = preprocessing(y)   
```

## Implementation of a fully connected regression network

```python
NN_model = Sequential()

# The Input Layer :
NN_model.add(Dense(256, kernel_initializer='normal',input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
NN_model.add(Dense(512, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(1024, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(512, kernel_initializer='normal',activation='relu'))

# The Output Layer :
NN_model.add(Dense(6, kernel_initializer='normal',activation='linear'))
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
NN_model.summary()

#from keras.optimizers import Adam
#adam =Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

NN_model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
```

```python
NN_model.fit(X_train, Y_train, epochs=200, batch_size=16, validation_split = 0.3)
```

# Image denoising with an hour Glass network(U-Net)

```python
from keras.layers import Input,UpSampling2D, concatenate
from keras.models import Model
from keras.optimizers import SGD
sgd = SGD(lr=0.0001,nesterov=True)
from keras.optimizers import Adam
adam =Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
inputs = Input(shape=(72,72,1))


#1st block

conv1a = Conv2D(16, 3, activation = 'relu',padding= 'same')(inputs)
conv1b = Conv2D(16, 3, activation = 'relu',padding= 'same')(conv1a)
pool1 = MaxPooling2D(pool_size=(2, 2),padding= 'same')(conv1b)

#2nd block
conv2a = Conv2D(32, 3, activation = 'relu',padding= 'same')(pool1)
conv2b = Conv2D(32, 3, activation = 'relu',padding= 'same')(conv2a)
pool2 = MaxPooling2D(pool_size=(2, 2),padding= 'same')(conv2b)

#3rd block
conv3a = Conv2D(64, 3, activation = 'relu',padding= 'same')(pool2)
conv3b = Conv2D(64, 3, activation = 'relu',padding= 'same')(conv3a)
pool3 = MaxPooling2D(pool_size=(2, 2),padding= 'same')(conv3b)

#4th block
conv4a = Conv2D(128, 3, activation = 'relu',padding= 'same')(pool3)
conv4b = Conv2D(128, 3, activation = 'relu',padding= 'same')(conv4a)

#UPSAMPLING
#5th block
up5 = UpSampling2D(size = (2,2))(conv4b)
up_conv5 = Conv2D(64, 2, activation = 'relu',padding= 'same')(up5)
merge5 = concatenate([conv3b,up_conv5],axis=3)
conv5b = Conv2D(64, 3, activation = 'relu',padding= 'same')(merge5)
conv5c = Conv2D(64, 3, activation = 'relu',padding= 'same')(conv5b)

#5th block
up6 = UpSampling2D(size = (2,2))(conv5c)
up_conv6 = Conv2D(32, 2, activation = 'relu',padding= 'same')(up6)
merge6 = concatenate([conv2b,up_conv6],axis=3)
conv6b = Conv2D(32, 3, activation = 'relu',padding= 'same')(merge6)
conv6c = Conv2D(32, 3, activation = 'relu',padding= 'same')(conv6b)

#5th block
up7 = UpSampling2D(size = (2,2))(conv6c)
up_conv7 = Conv2D(16, 2, activation = 'relu',padding= 'same')(up7)
merge7 = concatenate([conv1b,up_conv7],axis=3)
conv7b = Conv2D(16, 3, activation = 'relu', padding= 'same')(merge7)
conv7c = Conv2D(16, 3, activation = 'relu', padding= 'same')(conv7b)

#last layer
conv8 = Conv2D(1, 1, activation = 'relu',padding= 'same')(conv7c)

model = Model(input = inputs, output = conv8)


model.compile(loss='mean_squared_error',optimizer = adam)
model.summary()
```
